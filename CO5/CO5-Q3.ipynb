```python
#n grams in sentence analysis
import numpy as np
import pandas as pd
from sklearn import model_selection,naive_bayes,svm
```

```python
df=pd.read_csv('/content/sample_data/Reviews.csv',engine='python',error_bad_lines=False)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Score'].value_counts()
```

```python
df['postive ratings']=np.where(df['Score']>=3,1,0)
```

```python
df.head()
```

```python
df['postive ratings'].value_counts()
```

```python
#highly imbalanced
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df["Text"],df["postive ratings"],random_state=50)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vector=CountVectorizer(min_df=5,ngram_range=(1,2)).fit(x_train)
```

```python
vector.get_feature_names()
```

```python
len(vector.get_feature_names())
```

```python
x_train_vectorized=vector.transform(x_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_vectorized,y_train)
```

```python
pred=model.predict(vector.transform(x_test))
```

```python
print(pred)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
#auc score
from sklearn.metrics import roc_auc_score
print("AUC score is:",roc_auc_score(y_test,pred))
```
