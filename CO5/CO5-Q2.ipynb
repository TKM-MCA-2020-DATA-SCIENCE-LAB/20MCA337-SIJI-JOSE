```python
#webpage scraper
from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
webpage=requests.get("https://meesho.com/handbags-women/pl/6m6o8")
```

```python
webpage
```

```python
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
print(soup.prettify()) # used to nest tags
```

```python
names=soup.find_all('p',class_='Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS')
```

```python
names
```

```python
Bag_names=[]
for i in range(0,len(names)):
  Bag_names.append(names[i].get_text())
```

```python
Bag_names
```

```python
Price=soup.find_all('h5',class_='Text__StyledText-sc-oo0kvp-0 hiHdyy')
Price
```

```python
price_list=[]
for i in range(0,len(Price)):
  price_list.append(Price[i].get_text())
price_list
```

```python
import pandas as pd
df=pd.DataFrame()
```

```python
df['Names']=Bag_names
df['Price']=price_list
```

```python
df
```

```python
df.to_csv('Bags.csv',index=False)
```
