```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/sample_data/Heart_Disease_Prediction.csv')
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df.drop('Heart Disease',axis=1))
StandardScaler()
```

```python
df.columns[:-1]
```

```python
scaled_features=scaler.transform(df.drop('Heart Disease',axis=1))
scaled_features
```

```python
new_dataset=pd.DataFrame(scaled_features,columns=df.columns[:-1])
new_dataset
```

```python
from sklearn.model_selection import train_test_split
X=new_dataset
Y=df['Heart Disease']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,CategoricalNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
diseases=[]
```

```python
classifier=GaussianNB()
classifier.fit(X_train,Y_train)
Y_pred=classifier.predict(X_test)
diseases.append(accuracy_score(Y_test,Y_pred))
com=confusion_matrix(Y_test,Y_pred)
com
```
