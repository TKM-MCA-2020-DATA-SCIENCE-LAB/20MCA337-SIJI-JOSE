```python
#Simple Linear Regression
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/sample_data/Salary_Data.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,:-1].values
```

```python
# split the dataset into training set and testing set
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.3,random_state=30)
```

```python
# fit the training dataset
from sklearn.linear_model import LinearRegression
regsr=LinearRegression()
regsr.fit(x_train,y_train)
```

```python
y_pred=regsr.predict(x_test)
```

```python
#visualize
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regsr.predict(y_train),color='green')
plt.title("salary VS experience")
plt.xlabel("experience")
plt.ylabel("salary")
plt.show()
```

```python
# visualizing the test results
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,regsr.predict(y_test),color='green')
plt.title("salary VS experience")
plt.xlabel("experience")
plt.ylabel("salary")
plt.show()
```

```python
# Multiple Linear Regression
```

```python
ds=pd.read_csv('/content/sample_data/combined cycle powerplant.csv')
```

```python
ds.head()
```

```python
u=ds.drop('PE',axis=1).values
v=ds['PE'].values
```

```python
from sklearn.model_selection import train_test_split
u_train,u_test,v_train,v_test = train_test_split(u,v,test_size=0.3,random_state=45)
```

```python
#fitting
from sklearn.linear_model import LinearRegression
rgr=LinearRegression()
rgr.fit(u_train,v_train)
```

```python
v_pred=rgr.predict(u_test)
print(v_pred)
```

```python
#take the values of first row in u and compare our predicted v values with actual v value
rgr.predict([[23.64	,58.49,	1011.40,	74.20]])
```

```python
from sklearn.metrics import r2_score
r2_score(v_test,v_pred)
```

```python
# visualize 
plt.figure(figsize=(15,10))
plt.scatter(v_test,v_pred)
plt.title("actual VS predicted")
plt.xlabel("actual")
plt.ylabel("predicted")
```

```python
#print prdedicted values of our model
pred_ds=pd.DataFrame({'Actual value':v_test ,'Predicted value':v_pred})
pred_ds
```
