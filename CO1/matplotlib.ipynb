```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel('Velocity(m/s)')
plt.ylabel('Time(s)')
plt.title('Velocity-Time Graph')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g*-')
plt.show()
```

```python
plt.subplot(3,1,1)
plt.plot(x,y,'r--')
plt.subplot(3,1,2)
plt.plot(x,y,'g*-')
plt.subplot(3,1,3)
plt.plot(x,y,'b')
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.2,0.4,0.6,0.8])
axes.plot(x,y,'y')
axes.set_xlabel('Velocity')
axes.set_ylabel('Time')
axes.set_title('Velocity-Time Graph')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.2,0.4,0.6,0.8])
axes2=fig.add_axes([0.3,0.8,0.2,0.2])

axes1.plot(x,y,'r')
axes1.set_xlabel('Velocity')
axes1.set_ylabel('Time')
axes1.set_title('Velocity-Time Graph')

axes2.plot(x,y,'g')
axes2.set_xlabel('Distance')
axes2.set_ylabel('Time')
axes2.set_title('Distance-Time Graph')
```

```python
fig,axes=plt.subplots(1,2)
axes
```
